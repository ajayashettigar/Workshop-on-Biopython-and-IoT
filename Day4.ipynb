```python
#complement and reverse complement a sequence
from Bio.Seq import Seq

# Nucleotide sequence
org_seq = "ATGCGACTAG"

# Create a Seq object
seq_obj = Seq(org_seq)

#obtaining complement
comp_seq = seq_obj.complement()

#obtain reverse complement
recomp_seq = seq_obj.reverse_complement()

print(f"Original Sequence: {org_seq}")
print(f"Complementary Sequnce: {comp_seq}")
print(f"Reverse Complementary Sequence: {recomp_seq}")
```

# BLAST

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

#input seq file in fasta format
inp_seq = 'sample_sequence.fasta'

#read the input
query_seq = next(SeqIO.parse(inp_seq, 'fasta'))

#perform BLAST against ncbi db
result_handle = NCBIWWW.qblast('blastn', 'nt', query_seq.seq)

#save the BLAST results to a file
blast_result = 'blastn_results.xml'
with open(blast_result, 'w') as result_file:
    result_file.write(result_handle.read())
    
result_handle.close()

#parse the BLAST results
from Bio.Blast import NCBIXML

with open(blast_result,'r') as result_file:
    blast_record = NCBIXML.read(result_file)

#print info about BLAST results
print(f'Query: {blast_record.query[:50]}...')
print(f'Number of alignments: {len(blast_record.alignments)}')

#print info about the top alignment
if blast_record.alignments:
    top_alignment = blast_record.alignments[0]
    print(f"\nTop alignment:")
    print(f"Title: {top_alignment.title}")
    print(f"Length: {top_alignment.length}")
    print(f"E-value: {top_alignment.hsps[0].expect}")
    print(f"Identity: {top_alignment.hsps[0].identities}")
    print(f"Gaps: {top_alignment.hsps[0].gaps}")
    print(f"Query: {top_alignment.hsps[0].query[:50]}")
    print(f"Subject: {top_alignment.hsps[0].sbjct[:50]}")
```

# Single script for performing in a single step

```python
from Bio.Align.Applications import ClustalwCommandline
from Bio import Phylo, AlignIO
from Bio.Phylo.TreeConstruction import DistanceCalculator

#input sequence fasta file
input_file='ls_orchid.fasta'

#output aligned seq in clustalw format
output_clustalw_file='aligned_sequence.aln'

#output phylogenetic tree in newick format
output_tree_file='phylogenetic_tree.nwk'

#step1- perform MSA with clustalw
clustal_cline=ClustalwCommandline('clustalw2', infile=input_file, outfile=output_clustalw_file)
stdout, stderr=clustal_cline()

print(stdout)
print(stderr)

#step2- read aligned sequences
alignment=AlignIO.read(output_clustalw_file, 'clustal')

#step3- generate a phylogenetic tree
calculator=Phylo.TreeConstruction.DistanceCalculator('identity')
dm=calculator.get_distance(alignment)

constructor=Phylo.TreeConstruction.DistanceTreeConstructor(calculator)
nj_tree=constructor.nj(dm)

#step4- save thephylogenetic tree to a file
Phylo.write(nj_tree, output_tree_file, 'newick')

#step5- visualize the phylogenetic tree(optional)
#Phylo.draw(nj_tree)

#step6- initialize circos from phylogentic tree
tree_file=Phylo.read('phylogenetic_tree.nwk', 'newick')

circos, tv=Circos.initialize_from_tree(
    tree_file,
    r_lim=(30,100),
    leaf_label_size=5,
    line_kws=dict(color='green', lw=1.0),
)

#plot fig &set legend on center
fig=circos.plotfig()
fontsize=6,
loc="center",
bbox_to_anchor=(0.5,0.5),

fig.savefig('orchid.png')
fig.savefig('orchid.svg')
```

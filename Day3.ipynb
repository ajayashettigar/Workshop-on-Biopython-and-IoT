```python
#re.search()
import re

pattern = r'apple';
text = 'I have an apple';

match = re.search(pattern, text);
if match:
        print('pattern found:', match.group());
else:
    print('pattern not found');
```

```python
#re.match()
import re

pattern = r'apple';
text = 'I have an apple';

match = re.match(pattern, text);
if match:
        print('pattern found at the beginning:', match.group());
else:
    print('pattern not found at the beginning');
```

```python
#re.findall()
import re

pattern = r'\d+'; #Match one or more digits
pattern = r'[A-Za-z]' #Match all letters
pattern = r'and' #Match particular word
text = 'There are 10 apples and 20 oranges.';

matches = re.findall(pattern, text);
print('Matches:', matches)
```

```python
from scipy import misc

img = misc.face()
```

```python
type(img)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.imshow(img)
plt.show()
```

```python
from numpy import linalg #importing the linear algebre module from num
img_array = img/255
img_gray = img_array@[1000,1000,1000]

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
U, s, Vt = linalg.svd(img_gray)
U.shape, s.shape, Vt.shape
```

```python
import numpy as np

Sigma = np.zeros((U.shape[1], Vt.shape[0]))
np.fill_diagonal(Sigma, s)
linalg.norm(img_gray - U @ Sigma @ Vt)
k = 10
approx = U @ Sigma[:, :k] @ Vt[:k, :]
plt.imshow(approx, cmap='gray')
plt.show()
```
